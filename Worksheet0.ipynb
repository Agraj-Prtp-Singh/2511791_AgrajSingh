```python
#Exercise - Recursion:

#Task1

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total




nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

print(sum_nested_list(nested_list))



#Task2

def generate_permutation(s):
    if len(s) <= 1:
        return [s]
    
    result = []
    for i in range(len(s)):
        char = s[i]
        remaining = s[:i] + s[i+1:]

        for perm in generate_permutation(remaining):
            result.append(char + perm)
    
    # Remove duplicates if any
    result = list(set(result))
    return result

print(generate_permutation("abc"))



#Task3

directory_structure = {
"file1.txt": 200,
"file2.txt": 300,
"subdir1": {
"file3.txt": 400,
"file4.txt": 100
},
"subdir2": {
"subsubdir1": {
"file5.txt": 250
},
"file6.txt": 150
}
}

def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value,dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size

total = calculate_directory_size(directory_structure)
print(f"Total size of the directory: {total} KB")
        
#Task4
coins = [1,2,5]
amount = 11
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1,2,5],11))
print(min_coins([5,5,1],5))
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)
    ''' 
    Finds the longest common subsequence in the given strings using dynamic programming.

    Parameters:
        s1 (strings): First string
        s2 (stirngs): Second string

    Returns:
        int: The length longest common subsequence between two strings.
    '''

    memo = {}

    def rec(i, j):

        if i == 0 or j == 0:
            return 0
        
        if (i, j) in memo:
            return memo[(i,j)]
        
        if s1[i-1] == s2[j-1]:
            memo[(i,j)] = 1 + rec(i-1,j-1)

        else: 
            memo[(i,j)] = max(rec(i-1, j), rec(i, j-1))

        return memo[(i,j)]

    return rec(len(s1), len(s2))

print(longest_common_subsequence("abcd", "abc"))
```

```python
def knapsack_memo(weights, values, capacity):
    n = len(weights)
    memo = {}

    def rec(i, j):
        if i == 0 or j == 0:
            return 0
        
        if (i, j) in memo:
            
            return memo[(i,j)]
        
        if weights[i-1] <= j:
            memo[(i,j)] = max(rec(i-1,j), values[i-1]+rec(i-1, j - weights[i-1]))
        
        else:
            memo[(i,j)] = rec(i-1, j)
        
        return memo[(i,j)]
    
    return rec(n, capacity)

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack_memo(weights, values, capacity))
```
