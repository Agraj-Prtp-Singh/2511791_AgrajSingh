```python
#Problem - 1: Array Creation:
#1.Initializ an empty array with size 2x2
import numpy as np
arr_empty = np.empty([2,2])
print("\n",arr_empty)

#2. Initialize an all one array with size 4X2.
arr_ones = np.ones([4,2])
print("\n", arr_ones)

#3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
arr_full = np.full((3, 3), fill_value=2, dtype =int)
print("\n", arr_full)

#4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
array1 = np.array([[5, 6], [7, 8]])
arr_zeros_like = np.zeros_like(array1)
print("\n", arr_zeros_like)

#5 Return a new array of ones with the same shape and type as a given array.{hint: np.ones_like}
arr_ones_like = np.ones_like(array1)
print("\n", arr_ones_like)

#6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}
new_list = [1, 2, 3, 4]
arr_from_list = np.array(new_list)
print("\n", arr_from_list)
```

```python
#Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:
#1. Create an array with values ranging from 10 to 49. {Hint:np.arrange()}.
import numpy as np
arr = np.arange(10,50)
print(arr)

#2 Create a 3X3 matrix with values ranging from 0 to 8.{Hint:look for np.reshape()}
arr1 = np.arange(9).reshape(3,3)
print("\n",arr1)

#3. Create a 3X3 identity matrix.{Hint:np.eye()}
c = np.eye(3,3)
print("\n",c)

#4. Create a random array of size 30 and find the mean of the array.{Hint:check for np.random.random() and array.mean() function}
d = np.random.random(30)
print("\nRandom_Array:",d)
print("\nRandom_Array_Mean:",d.mean())

#5. Create a 10X10 array with random values and find the minimum and maximum values.
e = np.random.random((10,10))
print("/nMinimum Value: ", e.min())
print("/nMaximum Value: ", e.max())

#6. Create a zero array of size 10 and replace 5th element with 1.
f = np.zeros(10)
f[4] = 1
print("\n",f)

#7. Reverse an array arr = [1,2,0,0,4,0].
arr = [1,2,0,0,4,0]
reverse_arr = np.flip(arr)
print("Normal: ",arr)
print("Reverse: ",reverse_arr)


#8. Create a 2d array with 1 on border and 0 inside.
border_array = np.ones((4,4))
border_array[1:-1,1:-1] = 0
print("\nBorder array:\n", border_array)


#9. Create a 8X8 matrix and fill it with a checkerboard pattern.
h = np.zeros([8,8])
h[1::2, ::2] = 1
h[::2,1::2] = 1
print("\n",h)
```

```python
#Problem - 3: Array Operations:
import numpy as np
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

add_xy = x+y
print(add_xy)

sub_xy = y-x
print("\n",sub_xy)

mul_x = x * 3
print("\nMultiplying 'x' with 3: ",mul_x)

square_x = x ** 2
print("x squared =\n", square_x)

dot_vw = np.dot(v,w)
print("\nDot Product of v.w: ",dot_vw)
dot_xv = np.dot(x, v)
print("\nDot Product of x.v: ", dot_xv)

dot_xy = np.dot(x, y)
print("\nDot Product of x.y: ", dot_xy)


# 6. Concatenate x(and)y along row and Concatenate v(and)w along column. {Hint:try np.concatenate() or np.vstack() functions.

concat_xy_row = np.vstack((x, y))
print("\nx and y concatenated along row: \n", concat_xy_row)

concat_vw_column = np.hstack((v,w))
print("/nv and w concatenated along column: \n", concat_vw_column)

#7. Concatenate x(and)v; if you get an error, observe and explain why did you get the error?
xwithv = np.concatenate((x,v), axis=0)

#Explaination: NUmpy cannot concatenate arrays of different dimensions along default axis(0)
```

```python
#Problem - 4: Matrix Operations:
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

#1
I = np.eye(2, dtype = int)
A_inv = np.linalg.inv(A)
AdotA_inv = np.dot(A,A_inv)
print(AdotA_inv)
AdotA_inv_int = np.round(AdotA_inv).astype(int)
print("\n",AdotA_inv_int)

if np.array_equal(AdotA_inv_int,I):
    print("\nTrue")
else:
    print("\nFalse")

#2
AB = np.dot(A,B)
BA = np.dot(B,A)
print("AB=\n",AB)
print("BA=\n",BA)

if np.array_equal(AB, BA):
    print("\nBoth are equal")
else:
    print("\nNot equal")

#3
AB_T = np.dot(A,B).T
BT_AT = np.dot(B.T,A.T)
print("\n(AB)^T =\n", AB_T)
print("\nB^T A^T =\n", BT_AT)

if np.array_equal(AB_T, BT_AT):
    print("Property Verified: (AB)^T = B^T A^T")
else:
    print("Property Not Verified")


#4
A_eq = np.array([[2, -3, 1],[1, -1, 2],[3, 1, -1]])
B_eq = np.array([-1, -3, 9])

X = np.dot(np.linalg.inv(A_eq), B_eq)
print("\nSolution of [x, y, z] =", X)
```
