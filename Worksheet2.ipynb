```python
#Task2
import pandas as pd;
df = pd.read_csv(r"C:\Personal\Study Materials\AI\Week2\Dataset\medical_students_dataset.csv")
df.info()
null_count = df.isnull().sum()
print("\nMissing values per column:\n", null_count)
df["Age"].fillna(df["Age"].median(), inplace=True) #used median instead of mean as median works better if outliers exist
df["Gender"].fillna(df["Gender"].mode()[0], inplace = True) #Using mode to replace nan with most repeated category
df['Height'].fillna(df['Height'].median(), inplace=True)
df['Weight'].fillna(df['Weight'].median(), inplace=True)
df['BMI'].fillna(df['BMI'].median(), inplace=True)
df['Temperature'].fillna(df['Temperature'].mean(), inplace=True)   # stable
df['Heart Rate'].fillna(df['Heart Rate'].median(), inplace=True)
df['Blood Pressure'].fillna(df['Blood Pressure'].median(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].median(), inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Blood Type'].fillna(df['Blood Type'].mode()[0], inplace=True)
df['Diabetes'].fillna(df['Diabetes'].mode()[0], inplace=True)
df['Smoking'].fillna(df['Smoking'].mode()[0], inplace=True)
df["Student ID"].fillna(-1, inplace = True)
# df.isnull().sum()
df.duplicated().sum()
df = df.drop_duplicates()

print("\nMissing values after imputation:")
print(df.isnull().sum())

print("\nFinal shape after duplicate removal:")
print(df.shape)



```
